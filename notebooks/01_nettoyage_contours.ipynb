# 01 Nettoyage des contours des départements

## Objectifs

1. Téléchargment
2. Utilisation de geopandas
3. Essai de carte
3. Sauvegarde dans un format réutilisable

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1. Téléchargement

```python
url="https://france-geojson.gregoiredavid.fr/repo/departements.geojson"

df = gpd.read_file(url)

df.head()
```

```python
# test
with plt.style.context("dark_background"):
    df.plot("code", figsize=(10, 10))
    plt.axis("off")
```

## 2. Sauvegarde

```python
df.to_csv("../data_clean/clean_geography.csv", index=False)
```

```python
df = pd.read_csv("../data_clean/clean_geography.csv")
df = gpd.GeoDataFrame(
    df.loc[:, [c for c in df.columns if c != "geometry"]],
    geometry=gpd.GeoSeries.from_wkt(df["geometry"]),
    crs="epsg:3005",
)
```

```python
# test
with plt.style.context("dark_background"):
    df.plot("code", figsize=(10, 10))
    plt.axis("off")
```
