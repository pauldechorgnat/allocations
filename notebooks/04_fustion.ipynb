```python
import pandas as pd
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path_pop = "../data_clean/clean_population.csv"
path_geo = "../data_clean/clean_geography.csv"
path_alo = "../data_clean/clean_allocations.csv"
```

```python
df_pop = pd.read_csv(path_pop)
df_geo = pd.read_csv(path_geo)
df_alo = pd.read_csv(path_alo)
```

```python
df_pop["year"] = df_pop["year"].astype(str)
df_pop.head()
```

```python
df_geo.head()
```

```python
df_alo["year"] = pd.to_datetime(df_alo["month"]).dt.year.astype(str)
df_alo.head()
```

```python
df = pd.merge(
    left=df_alo,
    right=df_pop,
    on=["departement", "year"],
    how="left"
)

df.head()
```

```python
df.isna().sum()
```

```python
df = pd.merge(
    left=df,
    right=df_geo,
    how="left",
    on=["code"]
)

```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df[df["nom"].isna()]["departement"].unique()
```

```python
df = df.dropna(subset=["nom"])
```

```python
df.isna().sum()
```

```python
df.to_csv("../data_clean/master.csv", index=False)
```
